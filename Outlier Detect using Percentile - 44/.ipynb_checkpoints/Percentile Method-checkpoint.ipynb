```python
# Capping is called as Winsorization Technique
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('weight-height.csv')
df.head()
```

```python
df.shape
```

```python
df['Height'].describe()
```

```python
import seaborn as sns
```

```python
sns.distplot(df['Height'])
```

```python
sns.boxplot(df['Height'])
```

```python
upper_limit = df['Height'].quantile(0.99) # 99 percent se above 
upper_limit
```

```python
lower_limit = df['Height'].quantile(0.01) # 1 se below
lower_limit
```

```python
new_df = df[(df['Height'] <= 74.78) & (df['Height'] >= 58.13)]
```

```python
new_df['Height'].describe()
```

```python
df['Height'].describe()
```

```python
sns.distplot(new_df['Height'])
```

```python
sns.boxplot(new_df['Height'])
```

```python
# Capping --> Winsorization
df['Height'] = np.where(df['Height'] >= upper_limit,
        upper_limit,
        np.where(df['Height'] <= lower_limit,
        lower_limit,
        df['Height']))
```

```python
df.shape
```

```python
df['Height'].describe()
```

```python
sns.distplot(df['Height'])
```

```python
sns.boxplot(df['Height'])
```
